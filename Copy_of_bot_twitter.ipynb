```python
def log(log_text):
    timestamp = time.strftime("%Y%m%d%H%M%S")
    log_text = str(timestamp) + " ➾ " + log_text
    print(log_text)
    #log_file = open("log.txt", "a", encoding="utf-8")
    #log_file.write(log_text + "\n")
    #log_file.close()

    # Define the log file path
    log_file_path = "/content/drive/MyDrive/Colab Notebooks/bot-twitter-main/dist/twitter/log/log.txt"

    # Ensure the directory exists, create if not
    log_dir = os.path.dirname(log_file_path)
    if not os.path.exists(log_dir):
        os.makedirs(log_dir)
    # List all existing log files
    log_files = glob.glob(f"{log_dir}log_file_at_*.txt")

    # Delete the oldest log files if there are more than 5
    if len(log_files) >= 5:
        # Sort log files by creation time (oldest first)
        log_files.sort(key=os.path.getmtime)

        # Delete the oldest log file
        os.remove(log_files[0])
         # Define the log file path with dynamic timestamp
    log_file_path = f"{log_dir}log_file_at_{timestamp}.txt"

    # Write to the log file, creating a new file each time
    with open(log_file_path, "w", encoding="utf-8") as log_file:
        log_file.write(log_text + "\n")
```
